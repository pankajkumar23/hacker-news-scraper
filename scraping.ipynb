```python
from re import X
import requests
from bs4 import BeautifulSoup
url="https://news.ycombinator.com"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
titles = soup.find_all('span',class_="titleline")
scores = soup.find_all('span', class_='score')
news_list = []
for title,score in zip(titles[:10],scores[:10]):
      name = title.text
      link= title.find('a')['href']
      points = int(score.text.replace("points",""))
      news_list.append([name, link , points])
news_list.sort(key=lambda x : x[2] , reverse=True )

for news in news_list[:5]:
  print(news[2],"points",news[0])
```
